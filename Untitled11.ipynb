```python
#ran into issue where i wanted to take the lower bound of owners for each game

#defined a function to take thevariable split it into two strings then serpating it by'-'' 
#then it will create a new column called lower and upper bound for owners 
#to see if that amakes a differnce foir aanything in this dataset
```

```python
#define a function to calculate the upper bound

def calculate_upper_bound(owners_str):
    owners_range = owners_str.split("-") 
    upper_bound_str = owners_range[1] #range of numbers for owners the upper bound number
    upper_bound = int(upper_bound_str) #gets interger upper_bound
    return upper_bound

df["upper_bound"] = df["owners"].apply(calculate_upper_bound)

    
```

```python
#define a function to calculate the lower bound

#defines a function calculate_lower_bound 
def calculate_lower_bound(owners_str): #range of numbers of owners for the game in dataframe
    owners_range = owners_str.split("-") 
    lower_bound_str = owners_range[0] #range of numbers for owners the lower bound number
    lower_bound = int(lower_bound_str) # gets interger lower_bound
    return lower_bound

#
df["lower_bounds"] = df["owners"].apply(calculate_lower_bound)
#create a new column called"lower_bound" & "upper_bound"


#it worked 
```
